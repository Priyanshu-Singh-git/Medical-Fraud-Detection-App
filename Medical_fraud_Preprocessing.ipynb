```python
import pandas as pd


inpatient_df = pd.read_csv(r"F:\Medical_Fraud\Medical Invoice dataset\Train_In.csv")    
outpatient_df = pd.read_csv("F:\Medical_Fraud\Medical Invoice dataset\Train_Out.csv")  
beneficiary_df = pd.read_csv(r"F:\Medical_Fraud\Medical Invoice dataset\Train_Ben.csv")  
fraud_labels_df = pd.read_csv("F:\Medical_Fraud\Medical Invoice dataset\Train_label.csv")  

```

```python

combined_claims = pd.concat([inpatient_df, outpatient_df], ignore_index=True)
combined_data = combined_claims.merge(beneficiary_df, on="BeneID", how="left")
final_dataset = combined_data.merge(fraud_labels_df, on="Provider", how="left")


final_dataset.head()

```

```python
from sklearn.preprocessing import MinMaxScaler

num_cols = final_dataset.select_dtypes(include=["number"]).columns
final_dataset[num_cols] = final_dataset[num_cols].fillna(final_dataset[num_cols].median())


cat_cols = final_dataset.select_dtypes(include=["object"]).columns
final_dataset[cat_cols] = final_dataset[cat_cols].fillna(final_dataset[cat_cols].mode().iloc[0])




#final_dataset.to_csv("final_dataset_preprocessed.csv", index=False)


final_dataset.head()
```

```python
final_dataset = final_dataset.drop(columns = ["ClmProcedureCode_6"],axis=1)
```

```python
final_dataset.head()
```

```python
date_cols = ["AdmissionDt", "DischargeDt","ClaimStartDt", "ClaimEndDt","DOD","DOB"]   
final_dataset[date_cols] = final_dataset[date_cols].apply(lambda x: pd.to_datetime(x).view('int64') // 10**9) #Unix timestamp
final_dataset.head()
```

```python
# Initialize LabelEncoder
label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    final_dataset[col] = le.fit_transform(final_dataset[col])  # Transform the column
    label_encoders[col] = le  # Store the encoder for future use
```

```python
scaler = MinMaxScaler()
num_cols = ["AdmissionDt", "DischargeDt","ClaimStartDt", "ClaimEndDt","DOD","DOB","InscClaimAmtReimbursed","DeductibleAmtPaid","IPAnnualReimbursementAmt","IPAnnualDeductibleAmt","OPAnnualReimbursementAmt","OPAnnualDeductibleAmt"]
final_dataset[num_cols] = scaler.fit_transform(final_dataset[num_cols])
final_dataset[num_cols] = final_dataset[num_cols].round(6)
```
